```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
ac = pd.read_csv('ac_.csv')
loss = pd.read_csv('loss_.csv')
train_counter = list(range(36000,600001,12000))
```

```python
fig = plt.figure(figsize=(12, 8))
plt.plot(train_counter, loss['relu'][2:], color='blue')
plt.plot(train_counter, loss['sigmoid'][2:], color='red')
plt.plot(train_counter, loss['tanh'][2:], color='green')
plt.plot(train_counter, loss['leakyrelu'][2:], color='brown')
plt.legend(['relu', 'sigmoid', 'tanh', 'leakyrelu'], loc='upper right')
plt.xlabel('number of training examples seen')
plt.ylabel('CrossEntropyLoss')
plt.show()
```

```python
fig = plt.figure(figsize=(12, 8))
plt.plot(train_counter, ac['relu'][2:], color='blue')
plt.plot(train_counter, ac['sigmoid'][2:], color='red')
plt.plot(train_counter, ac['tanh'][2:], color='green')
plt.plot(train_counter, ac['leakyrelu'][2:], color='brown')
plt.legend(['relu', 'sigmoid', 'tanh', 'leakyrelu'], loc='lower right')
plt.xlabel('number of training examples seen')
plt.ylabel('test accuracy')
plt.show()
```
